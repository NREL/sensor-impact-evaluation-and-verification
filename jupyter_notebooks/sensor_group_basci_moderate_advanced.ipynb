```python
import pandas as pd
```

```python
temp = pd.read_csv('Suggested Sensor Sets.csv')
```

```python
temp = temp[['E+ Output Field', 'Sensor Set']]
```

```python
temp = temp.iloc[1:185]
```

```python
temp
```

```python
sensor_group_dict = temp.set_index('E+ Output Field').to_dict()['Sensor Set']
```

```python
# post processing

weather_list = ['AK_Fairbanks', 'FL_Miami', 'KY_Louisville', 'MN_Duluth', 'SAU_Riyadh', 'TN_Knoxville', 'VA_Richmond']

for weather in weather_list:
    temp = pd.read_csv(f'results/module_3/final_sensor_importance_{weather}.csv')
    temp['sensor group'] = temp['sensor'].map(sensor_group_dict)
    temp = temp[['sensor', 'sensor group', 'selected possibility']]
    temp.to_csv(f'results/module_3/postprocessed/final_sensor_importance_postprocessed_{weather}.csv', index = None)
```
